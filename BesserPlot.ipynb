```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.spatial import ConvexHull
import matplotlib.animation as anime
import matplotlib.patches as patch
```

```python
#1 task
df=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")
plt.figure(figsize=(18, 12))

cats=np.unique(df['category'])
colorr=[plt.cm.tab10(i/float(len(cats)-1)) for i in range(0, len(cats), 1)]

for i in range(0, len(cats), 1):
    plt.scatter(x='area', y='poptotal', data=df[df['category']==cats[i]], label=cats[i], c=colorr[i])

plt.xlim(0, 0.1)
plt.ylim(0, 90000)
plt.xlabel('Area')
plt.ylabel('Population')
plt.title('Scatterplot of Midwest Area vs Population')
plt.legend(loc=1)

plt.show()
```

```python
#2 task
df=pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")
plt.figure(figsize=(18, 18))
cats=np.unique(df['category'])
colls=[plt.cm.tab10(i/float(len(cats)-1)) for i in range(0, len(cats), 1)]

for i in range(0, len(cats), 1):
    cf=df[df['category']==cats[i]]
    plt.scatter(x=cf['area'], y=cf['poptotal'], color=colls[i], label=cats[i], s=cf['popdensity']/2.5, edgecolors='black')

#plt.fill([1, 7, 5, 1], [2, 3, 2, 4], color='g', alpha=0.3)
#plt.gca().set(xlabel='Area', ylabel='Population', xlim=(0, 10), ylim=(0, 10))
plt.gca().set(xlabel='Area', ylabel='Population', xlim=(0, 0.1), ylim=(0, 90000))

dd=df[df['state']=='IN']

p1=dd.iloc[0,:]
pk=[0]*dd.shape[0]

for i in range(0, dd.shape[0], 1):
    pk[i]=dd.iloc[i,:]

for i in range(0, dd.shape[0], 1):
    for j in range(0, dd.shape[0], 1):
        plt.fill([p1['area'], pk[i]['area'], pk[j]['area']], [p1['poptotal'], pk[i]['poptotal'], pk[j]['poptotal']], color='g', alpha=1)
        #plt.fill([p1['area'], p2['area'], p3['area']], [p1['poptotal'], p2['poptotal'], p3['poptotal']], color='g', alpha=1)
        

plt.fill()
plt.title('Bubble plot with encircling')
plt.legend(loc=1, labelspacing=1)
plt.show()
```

```python
df=pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
x = df.loc[:, ['mpg']]
df['mpg_z'] = (x - x.mean())/x.std()
#df['colors'] = ['red' if x < 0 else 'green' for x in df['mpg_z']]
df.sort_values(by='mpg_z', inplace=True, ascending=True)
#print(df)

plt.figure(figsize=(18,18))


df2=df[df['mpg_z']<=0]
plt.barh(y='cars', width='mpg_z', data=df2, color='red', alpha=0.3, height=0.3)
df1=df[df['mpg_z']>0]
plt.barh(y='cars', width='mpg_z', data=df1, color='green', alpha=0.3, height=0.3)

plt.title('Diverging bars of cars mileage')
plt.xticks(rotation=90, fontsize=12)
plt.ylabel('Model')
plt.xlabel('Mileage')

plt.grid(linestyle='--', alpha=0.2)
plt.show()
```

```python
#22
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")
print(df.head())

plt.figure(figsize=(18,18))
ar=np.unique(df['cyl'])
for i in range(0, len(ar), 1):
    sns.kdeplot(data=df[df['cyl']==ar[i]]['cty'], shade=True, label='Cyl='+str(ar[i]), alpha=0.8)

plt.ylim(0, 0.7)
plt.title('Density plot of city mileage by n_CYLL')
plt.legend(loc=1, frameon=False)
plt.show()
```

```python
#15
df_raw = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")
dfr2=df_raw.groupby(['manufacturer']).mean()
print(dfr2.head())
dfr2.sort_values(by='cty', inplace=True)

#df_raw.sort(values(by=''))


fig, ax=plt.subplots(1,1)
fig.set_size_inches(18,18)
ax.bar(x=range(0, dfr2.shape[0], 1), height='cty', data=dfr2, width=0.4, color='#AA0000')
ax.set_xticklabels(dfr2.index, horizontalalignment='right')
ax.set_xticks(range(0, dfr2.shape[0], 1))
ax.xaxis.set_tick_params(labelrotation=60, labelsize=14)

xf=dfr2.columns.get_loc('cty')
for i in range(0, dfr2.shape[0], 1):
    ax.text(x=i, y=dfr2.iat[i,xf], s=str(round(dfr2.iat[i,xf],1)), verticalalignment='bottom', horizontalalignment='center')


#Rekt
r1=patch.Rectangle(xy=(0.125,0.06), width=0.45, height=0.065, color='#EE0000', alpha=0.2)
r2=patch.Rectangle(xy=(0.575,0.06), width=0.325, height=0.065, color='#00EE00', alpha=0.2)
fig.add_artist(r1)
fig.add_artist(r2)

ax.set_ylabel('Miles Per Gallon')
plt.show()
```

```python
#11
df = pd.read_csv("https://github.com/selva86/datasets/raw/master/mtcars.csv")
df['mpg_z']=(df['mpg']-df['mpg'].mean())/df['mpg'].std()
df.sort_values(by='mpg_z', ascending=True, inplace=True)

fig, ax=plt.subplots(1,1)
fig.set_size_inches(19,19)
ax.barh(y=range(0, df.shape[0], 1), width='mpg_z', data=df, height=0.07, color='black')

fx=df.columns.get_loc('mpg_z')
for i in range(df.shape[0]):
    if (df.iat[i,fx]>=0):
        ax.text(x=df.iat[i,fx], y=i, s=str(round(df.iat[i,fx],2)), horizontalalignment='left', verticalalignment='center', color='green', fontsize=13)
    else:
        ax.text(x=df.iat[i,fx], y=i, s=str(round(df.iat[i,fx],2)), horizontalalignment='right', verticalalignment='center', color='red', fontsize=13)

ax.set_yticklabels(df['cars'])
ax.set_yticks(range(0, df.shape[0], 1))
ax.set_xticks(range(-2, 3, 1))
ax.set_xlim(-2.5,2.5)
ax.grid(alpha=0.4, linestyle='--')

plt.show()
```
